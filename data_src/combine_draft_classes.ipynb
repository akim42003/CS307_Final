## Combine All Draft Class CSVs into One CSV

```python
import pandas as pd

draft_class_21_df = pd.read_csv('data/21_draft_class.csv')
draft_class_20_df = pd.read_csv('data/20_draft_class.csv')
draft_class_19_df = pd.read_csv('data/19_draft_class.csv')
draft_class_18_df = pd.read_csv('data/18_draft_class.csv')
draft_class_17_df = pd.read_csv('data/17_draft_class.csv')
draft_class_16_df = pd.read_csv('data/16_draft_class.csv')
draft_class_15_df = pd.read_csv('data/15_draft_class.csv')
draft_class_14_df = pd.read_csv('data/14_draft_class.csv')
draft_class_13_df = pd.read_csv('data/13_draft_class.csv')
draft_class_12_df = pd.read_csv('data/12_draft_class.csv')
draft_class_11_df = pd.read_csv('data/11_draft_class.csv')
draft_class_10_df = pd.read_csv('data/10_draft_class.csv')
draft_class_09_df = pd.read_csv('data/09_draft_class.csv')
draft_class_08_df = pd.read_csv('data/08_draft_class.csv')
draft_class_07_df = pd.read_csv('data/07_draft_class.csv')
```

```python
# Combine all dataframes into one

draft_class_df = pd.concat([draft_class_21_df, draft_class_20_df, draft_class_19_df, draft_class_18_df, draft_class_17_df, 
    draft_class_16_df, draft_class_15_df, draft_class_14_df, draft_class_13_df, draft_class_12_df, 
    draft_class_11_df, draft_class_10_df, draft_class_09_df, draft_class_08_df, draft_class_07_df])

draft_class_df
```

```python
# Drop Rk Column

draft_class_df = draft_class_df.drop(['Rk'], axis=1)
```

```python
# Save to CSV

draft_class_df.to_csv('combined_draft_class_07_to_21.csv', index=False)
```
